```python
import numpy as np 
import pandas as pd
from IPython.display import display
athletes_df = pd.read_csv(r"sport_athletes.csv")
athletes_df
```

```python
# 1
athletes_df.groupby('Sport Type')[['Weight(kg)', 'Height(cm)', 'Wins']].max().rename(columns={'Weight(kg)': 'max_weight', 'Height(cm)': 'max_height', 'Wins': 'max_wins'})
```

```python
# 2
athletes_df.groupby('Sport Type')[['Weight(kg)']].min().sort_values(by='Weight(kg)')
```

```python
# 3
athletes_df.groupby('Sport Type')['Height(cm)'].nlargest(2)
```

```python
# 4.a
print('4.a')
display(athletes_df.groupby(['Country', 'Sport Type']).max(numeric_only=True))

print('\n4.b')
display(athletes_df.groupby(['Country', 'Sport Type']).max(numeric_only=True).loc[('USA', 'Basketball')])

print('\n4.c')
display(athletes_df.groupby(['Country', 'Sport Type']).max(numeric_only=True).xs(level = 'Sport Type', key = 'Soccer'))

print('\n4.d')
display(athletes_df[athletes_df['Sport Type'].isin(['Soccer', 'Tennis'])].groupby(['Country', 'Sport Type']).max(numeric_only=True))

print('\n4.e')
display(athletes_df.groupby(['Country', 'Sport Type']).max(numeric_only=True).swaplevel().sort_index(level=['Sport Type', 'Country'],\
                                                                                         ascending=[True, False]))

print('\n4.f - in seperate')
display(athletes_df.agg({'Wins': ['min', 'max', 'mean', 'std'], 'Height(cm)': ['min', 'max', 'mean', 'std']}).transpose())

print('\n4.f - in combination')
athletes_df.groupby(['Height(cm)', 'Wins']).agg({'Height(cm)': ['min', 'max', 'mean', 'std'], 'Wins': ['min', 'max', 'mean', 'std']})
```

```python
# 5
new_rows = pd.DataFrame({ 'Country': ['Canada', 'USA'], 'Sport Type': ['Hockey', 'Skiing'],
'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'], 'Height(cm)': [183, 172],
'Weight(kg)': [84, 68], 'Wins': [3, 4], 'Olympic Date': ['01/02/2002', '12/02/2010']}, index=[36, 37])
display(pd.concat([athletes_df, new_rows], axis=0)) 
```

```python
# 6
athletes_df = pd.read_csv(r"sport_athletes.csv")
athletes_df['Olympic Date'] = pd.to_datetime(athletes_df['Olympic Date'], format='%d/%m/%Y')
print('\nWins by year')
athletes_df.set_index('Olympic Date', inplace=True)
athletes_wins_per_year_df = athletes_df['Wins'].resample(rule='YE').sum()
display(athletes_wins_per_year_df)
print('\nonly pozitive wins')
display(athletes_wins_per_year_df[athletes_wins_per_year_df > 0])
print('\nreplace to nan')
athletes_wins_per_year_df.replace(0, np.nan, inplace=True)
display(athletes_wins_per_year_df)
print('\ndrop nan')
athletes_wins_per_year_df.dropna(inplace=True)
display(athletes_wins_per_year_df)
```

```python
# 7
smoking_status_df = pd.read_csv(r"athletes_smoking.csv")
print('\nonly players with smoking status')
display(pd.merge(athletes_df, smoking_status_df, on='Athlete Name', how='inner'))
print('\nall players with and without smoking status')
display(pd.merge(athletes_df, smoking_status_df, on='Athlete Name', how='left'))
```
